```python
cat_feat=data.drop(['elevation' ,'aspect','slope','horizontal_distance_to_hydrology','Vertical_Distance_To_Hydrology' ,'Horizontal_Distance_To_Roadways' ,'Horizontal_Distance_To_Fire_Points'],axis=1)
```

```python
plt.figure(figsize=(15,60))
plotnum=1
for column in num_feat:
    ax=plt.subplot(16,3,plotnum)
    sns.distplot(num_feat[column])
    plt.xlabel(column)
    plotnum+=1
plt.show
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_feat=scaler.fit_transform(num_feat)
```

```python
num_feat_scaled=pd.DataFrame(scaled_feat,columns=num_feat.columns)
```

```python
num_feat_scaled
```

```python
cat_feat['class']=cat_feat['class'].map({"Lodgepole_Pine" : 0, "Spruce_Fir" : 1,"Douglas_fir" :2,"Krummholz" : 3,"Ponderosa_Pine" : 4,"Aspen" :5,"Cottonwood_Willow" : 6})
```

```python
plt.figure(figsize=(10,5), facecolor='white')
sns.countplot(cat_feat['class'])
```

```python
final_data=pd.concat([num_feat_scaled,cat_feat],axis=1)
```

```python
final_data
```

```python
X = final_data.drop(['class'],axis=1)
y = data['class']
```
